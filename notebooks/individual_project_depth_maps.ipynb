```
#Data title format e.g 1_rect.bmp
drive_data_dir = "/content/drive/My Drive/bag1/_start_/"
data_dir = "/content/data/bag1/_start_/"
```

```
# Convert images to JPEG
import cv2
import os
#print(os.listdir(data_dir)[0:25])

if not os.path.exists(data_dir):
  os.makedirs(data_dir)

image_list = os.listdir(drive_data_dir) 
for index, img_name in enumerate(image_list): 
  name = drive_data_dir + img_name
  img = cv2.imread(name)
  out_name = data_dir + img_name
  cv2.imwrite(out_name.replace('.bmp', '.jpg'), img)
  print(f'Image {index+1}/{len(image_list)} conversion completed')
```
